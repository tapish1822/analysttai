Part 1
In this assignment you are required to scrape all products from this URL:
https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2
C283&ref=sr_pg_1
Need to scrape atleast 20 pages of product listing pages
Items to scrape
• Product URL
• Product Name
• Product Price
• Rating
• Number of reviews

```python
import requests
from bs4 import BeautifulSoup

# Set the number of pages to scrape
num_pages = 20

# URL of the product listings page
url = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{}"

# Create lists to store the scraped data
product_urls = []
product_names = []
product_prices = []
ratings = []
review_counts = []

# Scrape data from each page
for page in range(1, num_pages + 1):
    # Get the page content
    page_url = url.format(page)
    response = requests.get(page_url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all product listings
    listings = soup.find_all("div", class_="sg-col-inner")

    # Extract data from each product listing
    for listing in listings:
        # Extract product URL
        link = listing.find("a", class_="a-link-normal a-text-normal")
        if link:
            product_urls.append("https://www.amazon.in" + link["href"])
        else:
            product_urls.append("")

        # Extract product name
        name = listing.find("span", class_="a-size-base-plus a-color-base a-text-normal")
        if name:
            product_names.append(name.text.strip())
        else:
            product_names.append("")

        # Extract product price
        price = listing.find("span", class_="a-price-whole")
        if price:
            product_prices.append(price.text.strip())
        else:
            product_prices.append("")

        # Extract product rating
        rating = listing.find("span", class_="a-icon-alt")
        if rating:
            ratings.append(rating.text.strip())
        else:
            ratings.append("")

        # Extract number of reviews
        review = listing.find("span", class_="a-size-base")
        if review:
            review_counts.append(review.text.strip())
        else:
            review_counts.append("")

# Print the scraped data
for i in range(len(product_urls)):
    print("Product URL:", product_urls[i])
    print("Product Name:", product_names[i])
    print("Product Price:", product_prices[i])
    print("Rating:", ratings[i])
    print("Number of Reviews:", review_counts[i])
    print("-------------------------")
```

Part 2
With the Product URL received in the above case, hit each URL, and add below items:
• Description
• ASIN
• Product Description
• Manufacturer
Need to hit around 200 product URL’s and fetch various information.
The entire data needs to be exported in a csv format

```python
import requests
from bs4 import BeautifulSoup
import csv

# List of product URLs
product_urls = [
    "https://www.amazon.in/dp/ASIN1",
    "https://www.amazon.in/dp/ASIN2",
    "https://www.amazon.in/dp/ASIN3",
    # Add more product URLs here
]

# Create a CSV file to store the scraped data
csv_file = open("product_data.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["ASIN", "Product Name", "Description", "Product Description", "Manufacturer"])

# Scrape data from each product URL
for url in product_urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract ASIN
    asin = url.split("/dp/")[1].split("/")[0]

    # Extract Product Name
    product_name_element = soup.find("span", id="productTitle")
    product_name = product_name_element.get_text(strip=True) if product_name_element else ""

    # Extract Description
    description_element = soup.find("div", id="feature-bullets")
    if description_element:
        description_items = description_element.find_all("span", class_="a-list-item")
        description = "\n".join([item.get_text(strip=True) for item in description_items])
    else:
        description = ""

    # Extract Product Description
    product_description_element = soup.find("div", id="productDescription")
    product_description = product_description_element.get_text(strip=True) if product_description_element else ""

    # Extract Manufacturer
    manufacturer_element = soup.find("a", id="bylineInfo")
    manufacturer = manufacturer_element.get_text(strip=True) if manufacturer_element else ""

    # Write the scraped data to the CSV file
    csv_writer.writerow([asin, product_name, description, product_description, manufacturer])

# Close the CSV file
csv_file.close()
```
